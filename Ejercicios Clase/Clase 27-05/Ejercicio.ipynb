```python
# Dataset con filtro por edad y red social más usada
print("Punto 3: Nuevo dataset con filtro de estudiantes menores o iguales a 20 años y que su red social preferida sea instagram \n")
df_redes_filtred = df_redes[(df_redes['Age'] <= 20) & (df_redes['Most_Used_Platform'] == 'Instagram')]

df_redes_filtred.head()

```

```python
# Dataset con ordenamiento por genero y edad
colum_order = ['Gender','Age'] # Se crea lista con lo campos a ordenar
tip_orden = [True,False] # Se crea lista con el indicador de ordenamiento asc o desc
print("Punto 4: Nuevo dataset con orde por genero y edad de manera descendente\n")
df_redes.sort_values(by= colum_order ,ascending=tip_orden)
```
